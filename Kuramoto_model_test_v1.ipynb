# Numerical simulation of Kuramoto oscillator on graph.

```python
# import libraries.
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
# self-made module
from Kuramoto_model_v1 import Kuramoto_model
```

test

```python
sys1 = Kuramoto_model()
sys1.set_const(omega_cnst = 15, K_cnst=0.1)

nx.draw_networkx(sys1.graph)
plt.show()

Phases, Ts = sys1.solve(0.01, 2000)
abs_op, phase_op =sys1.get_order_parameter()
```

```python
for phases in Phases:
    plt.plot(Ts, phases)
```

```python
plt.plot(Ts, abs_op)
plt.ylim(0, 1)
plt.title("abs of order parameter")
plt.show()
```

```python
ang_vel = sys1.get_angular_velocity()
```
